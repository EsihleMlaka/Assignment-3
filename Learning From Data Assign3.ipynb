•	Skewness for SAT = 0.1864347050585477: The positive skewness of 0.19 indicates that the distribution of SAT scores is slightly skewed to the right. 
•	This means that there are a few more individuals with high SAT scores compared to those with low scores, and the distribution has a longer tail on the right side.

•	Skewness for Average score (grade) = 0.002229824655558406: The skewness of 0.002 for the distribution of average scores indicates that the distribution is almost perfectly symmetric. This means that there are almost an equal number of individuals with scores above and below the mean, and the distribution has roughly the same length of tails on both sides.

8.	Standard Dev:
•	For Age, the SD is 4.261158274618215. This suggests that the ages in the dataset are relatively tightly clustered around the mean age. In other words, there is not a lot of variability in age among the individuals in the dataset.

•	For SAT, the SD is 274.48754523471183. This suggests that the SAT scores in the dataset are spread out over a wide range. There is a lot of variability in the scores, with some individuals scoring much higher or lower than the mean.

•	For Average score (grade), the SD is 11.408664391039792. This suggests that the grades in the dataset are somewhat spread out, but not as much as the SAT scores. Many students scored relatively close to the mean grade.

•	For Newspaper readership (times/wk), the SD is 1.4085282758282187. This suggests that the readership in the dataset is not spread out much at all. Most individuals read the newspaper a similar number of times per week, with very little variability in the data.

•	Kurtosis for normal distribution:
•	For the variable "Age", the kurtosis value is -0.4279, which is less than 0. This indicates that the distribution of ages is  slightly flatter than the normal distribution, also known as "platykurtic". This means that the tails of the distribution   are thinner than the tails of the normal distribution, indicating that there are fewer extreme values in the distribution.

•	For the variable "SAT", the kurtosis value is -0.9393, which is also less than 0. This indicates that the distribution of SAT scores is slightly flatter than the normal distribution, or "platykurtic". This means that the distribution has fewer extreme values than the normal distribution, similar to the distribution of ages.

•	For the variable "Average score (grade)", the kurtosis value is -1.1516, which is also less than 0. This indicates that the distribution of average scores is even flatter than the normal distribution, or "platykurtic". This means that the distribution has even fewer extreme values than the previous two distributions, and the tails are even thinner.


• For the variable "Newspaper readership (times/wk)", the kurtosis value is -1.3195, which is less than 0. This indicates that the distribution of newspaper readership is also flatter than the normal distribution, or "platykurtic". This means that the distribution has even fewer extreme values than the previous three distributions, and the tails are even thinner.


9.	Geometric Mean:
•	In this context, the geometric mean for age (24.084), SAT (1821.26), and average grade score (78.79) all represent the typical value that reflects the overall growth or performance of a particular population. For example, the geometric mean for SAT scores suggests that the typical student's score is 1821.26, which is more representative of the center of the distribution than the arithmetic mean, especially when there are extreme values in the data set.

•	For the geometric mean of newspaper readership, it appears to be 4.795, indicating that this is the typical frequency of reading a newspaper per week among the population under consideration. This measure can be useful in marketing or advertising research to determine the level of engagement or reach of a particular newspaper or publication.
